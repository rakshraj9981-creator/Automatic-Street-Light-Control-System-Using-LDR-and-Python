```python
import time

THRESHOLD = 300  # Light intensity threshold

def read_ldr_value():
    """
    Simulates reading LDR sensor value.
    In real hardware, this comes from ADC.
    """
    return int(input("Enter LDR value (0 - 1023): "))

while True:
    ldr_value = read_ldr_value()

    if ldr_value < THRESHOLD:
        print("Status: DARK | Street Light: ON")
    else:
        print("Status: BRIGHT | Street Light: OFF")

    time.sleep(2)
```
